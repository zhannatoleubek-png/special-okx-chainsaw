```python
!pip install requests pandas numpy ta --quiet

import requests
import pandas as pd
import numpy as np
import time
from datetime import datetime, timedelta
from ta.trend import EMAIndicator, MACD
from ta.momentum import RSIIndicator

# === üîê TELEGRAM ===
TELEGRAM_TOKEN = "8294892098:AAFX0Zzq9yN1on6UlID8f7vzif4dWR_7uWs"
CHAT_ID = "381202205"

# === üì° OKX API ===
BASE_URL = "https://www.okx.com"
HEADERS = {"User-Agent": "Mozilla/5.0"}

INTERVALS = ["5m", "15m", "1H", "4H"]
SLEEP_MINUTES = 10
REPORT_INTERVAL_HOURS = 6

last_signals = {}       # –ó–∞–ø–æ–º–∏–Ω–∞–µ—Ç –ø–æ—Å–ª–µ–¥–Ω–∏–µ —Å–∏–≥–Ω–∞–ª—ã
last_report_time = datetime.utcnow()

# === üí¨ Telegram ===
def send_telegram(message):
    url = f"https://api.telegram.org/bot{TELEGRAM_TOKEN}/sendMessage"
    data = {"chat_id": CHAT_ID, "text": message, "parse_mode": "HTML"}
    try:
        requests.post(url, data=data)
    except Exception as e:
        print("–û—à–∏–±–∫–∞ –æ—Ç–ø—Ä–∞–≤–∫–∏ –≤ Telegram:", e)

# === üìä –ó–∞–≥—Ä—É–∑–∫–∞ —Å–ø–∏—Å–∫–∞ –º–æ–Ω–µ—Ç ===
def get_symbols():
    try:
        url = f"{BASE_URL}/api/v5/public/instruments?instType=SWAP"
        data = requests.get(url, headers=HEADERS).json().get("data", [])
        return [d["instId"] for d in data if "USDT" in d["instId"]]
    except Exception as e:
        print("–û—à–∏–±–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ –º–æ–Ω–µ—Ç:", e)
        return []

# === üïØ –ó–∞–≥—Ä—É–∑–∫–∞ —Å–≤–µ—á–µ–π ===
def get_klines(symbol, interval="15m", limit=200):
    try:
        url = f"{BASE_URL}/api/v5/market/candles?instId={symbol}&bar={interval}&limit={limit}"
        r = requests.get(url, headers=HEADERS, timeout=10).json()
        df = pd.DataFrame(r["data"], columns=["ts","o","h","l","c","v","vCcy","vUsd","confirm"])
        df = df.astype({"o":float,"h":float,"l":float,"c":float})
        df = df.iloc[::-1].reset_index(drop=True)
        return df
    except:
        return None

# === üìà –ê–Ω–∞–ª–∏–∑ –º–æ–Ω–µ—Ç—ã ===
def analyze_symbol(symbol):
    tf_signals = []
    for tf in INTERVALS:
        df = get_klines(symbol, tf)
        if df is None or len(df) < 50:
            continue

        ema_fast = EMAIndicator(df["c"], 12).ema_indicator()
        ema_slow = EMAIndicator(df["c"], 26).ema_indicator()
        macd = MACD(df["c"]).macd()
        macd_signal = MACD(df["c"]).macd_signal()
        rsi = RSIIndicator(df["c"], 14).rsi()

        ema_diff = ema_fast.iloc[-1] - ema_slow.iloc[-1]
        macd_hist = macd.iloc[-1] - macd_signal.iloc[-1]
        rsi_val = rsi.iloc[-1]

        score = 0
        if ema_diff > 0: score += 1
        if macd.iloc[-1] > macd_signal.iloc[-1]: score += 1
        if 40 < rsi_val < 70: score += 1
        if ema_diff < 0: score -= 1
        if macd.iloc[-1] < macd_signal.iloc[-1]: score -= 1
        if rsi_val > 70 or rsi_val < 30: score -= 1

        tf_signals.append(score)

    if not tf_signals:
        return None

    avg_signal = np.mean(tf_signals)
    agree = sum(np.sign(tf_signals))

    # === –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ç–∏–ø —Å–∏–≥–Ω–∞–ª–∞ ===
    if avg_signal >= 1.5 and agree > 2:
        sig = "üöÄ –°–ò–õ–¨–ù–´–ô –õ–û–ù–ì"
    elif 0.5 <= avg_signal < 1.5 and agree > 1:
        sig = "üìà –õ–û–ù–ì"
    elif -1.5 <= avg_signal <= -0.5 and agree < -1:
        sig = "üìâ –®–û–†–¢"
    elif avg_signal < -1.5 and agree < -2:
        sig = "üí£ –°–ò–õ–¨–ù–´–ô –®–û–†–¢"
    else:
        return None

    df = get_klines(symbol, "15m")
    if df is None: return None

    atr = (df["h"].iloc[-14:].mean() - df["l"].iloc[-14:].mean())
    price = df["c"].iloc[-1]
    if "–õ–û–ù–ì" in sig:
        sl = price - 2 * atr
        tp = price + 3 * atr
    else:
        sl = price + 2 * atr
        tp = price - 3 * atr

    return {"symbol": symbol, "signal": sig, "price": round(price, 4), "sl": round(sl, 4), "tp": round(tp, 4)}

# === üì¢ –û—Ç—á—ë—Ç –∫–∞–∂–¥—ã–µ 6 —á–∞—Å–æ–≤ ===
def send_report(all_signals):
    long_count = sum(1 for s in all_signals if "–õ–û–ù–ì" in s["signal"])
    short_count = sum(1 for s in all_signals if "–®–û–†–¢" in s["signal"])
    msg = (
        f"üïï <b>6-—á–∞—Å–æ–≤–æ–π –æ—Ç—á—ë—Ç:</b>\n\n"
        f"–í—Å–µ–≥–æ —Å–∏–≥–Ω–∞–ª–æ–≤: {len(all_signals)}\n"
        f"üìà –õ–æ–Ω–≥–æ–≤: {long_count}\n"
        f"üìâ –®–æ—Ä—Ç–æ–≤: {short_count}\n"
        f"‚è∞ –í—Ä–µ–º—è: {datetime.utcnow().strftime('%Y-%m-%d %H:%M:%S')} UTC"
    )
    send_telegram(msg)

# === üîÅ –ì–ª–∞–≤–Ω—ã–π —Ü–∏–∫–ª ===
def run_bot():
    global last_signals, last_report_time

    print("ü§ñ –ë–æ—Ç –∑–∞–ø—É—â–µ–Ω. –ê–Ω–∞–ª–∏–∑ –∫–∞–∂–¥—ã–µ 10 –º–∏–Ω—É—Ç.")
    send_telegram("ü§ñ –ë–æ—Ç –∑–∞–ø—É—â–µ–Ω. –ê–Ω–∞–ª–∏–∑ –∫–∞–∂–¥—ã–µ 10 –º–∏–Ω—É—Ç.")

    symbols = get_symbols()
    if not symbols:
        print("‚ùå –û—à–∏–±–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ —Å–ø–∏—Å–∫–∞ –º–æ–Ω–µ—Ç.")
        return

    while True:
        start = datetime.utcnow().strftime("%H:%M:%S")
        print(f"\nüïì {start} ‚Äî –Ω–∞—á–∞–ª–æ –∞–Ω–∞–ª–∏–∑–∞ {len(symbols)} –º–æ–Ω–µ—Ç")

        new_signals = []
        for i, sym in enumerate(symbols[:80]):  # –æ–≥—Ä–∞–Ω–∏—á–∏–º –¥–ª—è —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏
            result = analyze_symbol(sym)
            if not result:
                continue

            key = f"{result['symbol']}_{result['signal']}"
            if key == last_signals.get(result["symbol"]):
                continue  # —Å–∏–≥–Ω–∞–ª –Ω–µ –∏–∑–º–µ–Ω–∏–ª—Å—è ‚Äî –ø—Ä–æ–ø—É—Å–∫–∞–µ–º

            new_signals.append(result)
            last_signals[result["symbol"]] = key

            time.sleep(1)

        # === –û—Ç–ø—Ä–∞–≤–∫–∞ —Å–∏–≥–Ω–∞–ª–æ–≤ ===
        if new_signals:
            strong = [s for s in new_signals if "–°–ò–õ–¨–ù–´–ô" in s["signal"]]
            normal = [s for s in new_signals if "–°–ò–õ–¨–ù–´–ô" not in s["signal"]]

            if strong:
                send_telegram(f"üî• <b>{len(strong)} –°–ò–õ–¨–ù–´–• –°–ò–ì–ù–ê–õ–û–í:</b>")
                for s in strong:
                    msg = f"<b>{s['symbol']}</b>\n{s['signal']}\nüí∞ {s['price']}\nüõë SL: {s['sl']}\nüéØ TP: {s['tp']}"
                    send_telegram(msg)
                    time.sleep(1.5)

            if normal:
                send_telegram(f"üìä <b>{len(normal)} —Å–∏–≥–Ω–∞–ª–æ–≤:</b>")
                for s in normal:
                    msg = f"<b>{s['symbol']}</b>\n{s['signal']}\nüí∞ {s['price']}\nüõë SL: {s['sl']}\nüéØ TP: {s['tp']}"
                    send_telegram(msg)
                    time.sleep(1.5)
        else:
            print("üì≠ –ù–æ–≤—ã—Ö —Å–∏–≥–Ω–∞–ª–æ–≤ –Ω–µ—Ç.")

        # === –û—Ç—á—ë—Ç –∫–∞–∂–¥—ã–µ 6 —á–∞—Å–æ–≤ ===
        if datetime.utcnow() - last_report_time >= timedelta(hours=REPORT_INTERVAL_HOURS):
            send_report(list(last_signals.values()))
            last_report_time = datetime.utcnow()

        print(f"‚è≥ –°–ª–µ–¥—É—é—â–∏–π –∞–Ω–∞–ª–∏–∑ —á–µ—Ä–µ–∑ {SLEEP_MINUTES} –º–∏–Ω—É—Ç...\n")
        for _ in range(SLEEP_MINUTES * 60):
            time.sleep(1)

# === ‚ñ∂Ô∏è –ó–∞–ø—É—Å–∫ ===
run_bot()
```
